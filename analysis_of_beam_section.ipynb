```python
#define node location and support condition
l = 25 #beam length in m
```

```python
import planesections as ps
```

```python
#lets instaniate or create a beam object using newEuluerbeam method passing the length of the beam as an argument
```

```python
beam = ps.newEulerBeam2D(l) #we are importing planesection library
```

```python
#define beam and support
# support = { free = [0,0,0], roller =[0,1,0], pinned = [1,1,0], fixed =[1,1,1]}
```

```python
pinned = [1,1,0]
beam.setFixity(0, pinned)
beam.setFixity(l*0.8, pinned)
```

```python
#define point load and label
```

```python
p = -1000
beam.addLabel(0, label ='A')
beam.addLabel(10, label = 'E')
beam.addVerticalLoad(15, 2*p, label = 'D')
beam.addLabel(20, label= 'B')
beam.addVerticalLoad(25, 3*p, label = 'C')
```

```python
#define distributed loads
beam.addDistLoadVertical(0, l*0.4, p)
```

```python
#draw the beam diagram
```

```python
ps.plotBeamDiagram(beam)
```

```python
#run the analysis
analysis = ps.OpenSeesAnalyzer2D(beam)
```

```python
analysis.runAnalysis()
```

```python
#plot the shearforce and bending moment
ps.plotShear2D(beam, scale = 0.0002, yunit = 'KN')
ps.plotMoment2D(beam, scale = 0.002, yunit = 'kNm')
```
